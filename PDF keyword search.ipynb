```python
import pdfplumber
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords
```

```python
# Define the PDF file path and keywords to search for
pdf_file_path = '/content/HPP_2021-2022_NRW-HCC_BurnSurgeAnnex_Final.pdf'
keywords = ["Indicators", "triggers", "alerts", "notifications", "situational", "impact", "specialty",
"resources", "situational awareness", "Define the impact", "Determine specialty population needs",
""
"Document resource gaps", "external needs",
"Subject Matter", "Expert groups",
"Patient transfer",
"transport prioritization",
"Just in time (JIT) training", "JIT",
"plan for specialty functions",
"Local risks for MCI due to burn",
"Medical supplies for burn",
"American Burn Association", "coordination",
"Critical Care resources" ,
"ground", "air",
]

# Initialize a counter for each keyword
keyword_counts = {keyword: 0 for keyword in keywords}

# Open the PDF file
with pdfplumber.open(pdf_file_path) as pdf:
    # Iterate through each page in the PDF
    for page in pdf.pages:
        # Extract text from the page
        page_text = page.extract_text()

        # Search for each keyword in the page text and update the counts
        for keyword in keywords:
            keyword_counts[keyword] += page_text.lower().count(keyword.lower())

# Print the counts for each keyword
for keyword, count in keyword_counts.items():
    print(f'{keyword}: {count} times')
```

```python
def extract_text_from_pdf(pdf_file):
    text = ""
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

def get_most_used_words(text, n=10):
    # Tokenize the text into words using regular expressions
    words = re.findall(r'\w+', text.lower())

    # Count the word frequencies
    word_counts = Counter(words)

    # Get the n most common words
    most_common_words = word_counts.most_common(n)

    return most_common_words

if __name__ == "__main__":
    pdf_file =  '/content/HPP_2021-2022_NRW-HCC_BurnSurgeAnnex_Final.pdf' # Replace with the path to your PDF file
    text = extract_text_from_pdf(pdf_file)
    most_common_words = get_most_used_words(text, n=30)

    for word, frequency in most_common_words:
        print(f"{word}: {frequency}")
```
