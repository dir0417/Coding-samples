```python
import openpyxl
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Load your Excel file
workbook = openpyxl.load_workbook('/content/Contacts.xlsx')
sheet = workbook.active

# Email server settings
smtp_server = 'smtp.gmail.com'
smtp_port = 587
smtp_username = 'diana.ir@state.co.us'
smtp_password = '3dw4rd35cdphe@@'

# Create an email message
subject = 'EMResource account deactivation'
from_email = 'diana.ir@state.co.us'
email_message = '''
Hello {name},

test test test test test test test

Diana Ir
Systems Data Analyst
Office of Emergency Preparedness and Response (OEPR)
Division of Disease Control and Public Health Response (DCPHR)
Colorado Department of Public Health and Environment (CDPHE)
4300 Cherry Creek Drive South, Denver, CO 80246-1530
Phone number: (720) 432 - 4658
Email: diana.ir@state.co.us
'''

# Connect to the SMTP server
server = smtplib.SMTP(smtp_server, smtp_port)
server.starttls()
server.login(smtp_username, smtp_password)

# Iterate through the spreadsheet and send emails
for row in sheet.iter_rows(min_row=2, values_only=True):
    to_email = row[0]  # Assuming email addresses are in the first column
    name = row[1]  # Assuming names are in the second column

    # Create a message container
    msg = MIMEMultipart()
    msg['From'] = from_email
    msg['To'] = to_email
    msg['Subject'] = subject

    # Add the email message with the recipient's name
    email_text = email_message.format(name=name)
    msg.attach(MIMEText(email_text, 'plain'))

    # Send the email
    server.sendmail(from_email, to_email, msg.as_string())

# Quit the SMTP server
server.quit()
```
