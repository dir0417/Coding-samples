```python
import fitz  # PyMuPDF
```

```python
def search_and_extract_pdf(pdf_path, keyword, context_size=5000):
    doc = fitz.open(pdf_path)

    for page_num in range(doc.page_count):
        page = doc[page_num]
        text = page.get_text()
        keyword_occurrences = []

        # Find all occurrences of the keyword in the text
        start_index = text.find(keyword)
        while start_index != -1:
            end_index = start_index + len(keyword)
            keyword_occurrences.append((start_index, end_index))
            start_index = text.find(keyword, end_index)

        # Display context around each keyword occurrence
        for start, end in keyword_occurrences:
            start_context = max(0, start - context_size)
            end_context = min(len(text), end + context_size)
            context = text[start_context:end_context]
            print(f"Page {page_num + 1}, Keyword Occurrence: '{context}'")

    doc.close()

# Example usage
pdf_path = '/content/HPP_2021-2022_BurnSurgeAnnex_SCHCC_FINAL.pdf'
search_keyword = 'coordination'
output_file_path = 'output.txt'

search_and_extract_pdf(pdf_path, search_keyword, output_file_path)

```
