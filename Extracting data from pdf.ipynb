```python
import tabula
import pandas as pd
from openpyxl import Workbook
```

```python
# Replace 'input.pdf' with the path to your PDF file
pdf_path = '/content/CDPHE OEPR Supply Chain Integrity Assessment Summary Report - FINAL.pdf'

# Extract tables from the PDF as DataFrame objects
tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)
```

```python
# Create a new Excel workbook
excel_workbook = Workbook()

# Iterate through the extracted tables
for idx, table_df in enumerate(tables):
    # Create a new Excel worksheet
    excel_worksheet = excel_workbook.create_sheet(title=f"Table_{idx+1}")

    # Convert the DataFrame to a list of lists
    table_data = table_df.values.tolist()

    # Write the table data to the Excel worksheet
    for row in table_data:
        excel_worksheet.append(row)

# Remove the default sheet created by openpyxl
default_sheet = excel_workbook['Sheet']
excel_workbook.remove(default_sheet)

# Replace 'output.xlsx' with the desired output Excel file name
excel_output_path = 'output.xlsx'

# Save the Excel workbook
excel_workbook.save(excel_output_path)
```
