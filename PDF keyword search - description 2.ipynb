```python
import fitz  # PyMuPDF
```

```python
def search_keyword_in_pdf(pdf_path, keyword):
    doc = fitz.open(pdf_path)

    for page_num in range(doc.page_count):
        page = doc[page_num]
        text = page.get_text()

        if keyword.lower() in text.lower():
            # Display context around the keyword
            context_start = max(0, text.lower().find(keyword.lower()) - 50)
            context_end = min(len(text), text.lower().find(keyword.lower()) + 50)
            context = text[context_start:context_end]

            print(f"Page {page_num + 1}, Context: ...{context}...")

    doc.close()

if __name__ == "__main__":
    pdf_path = "/content/HPP_2021-2022_BurnSurgeAnnex_SCHCC_FINAL.pdf"
    keyword = "coordination"

    search_keyword_in_pdf(pdf_path, keyword)
```
