```python
import fitz  # PyMuPDF
```

```python
def search_keyword_in_pdf(pdf_path, keyword):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)

    # Initialize a list to store page numbers where the keyword is found
    keyword_pages = []

    # Iterate through each page in the PDF
    for page_number in range(pdf_document.page_count):
        # Get the page
        page = pdf_document[page_number]

        # Search for the keyword in the text of the page
        keyword_instances = page.search_for(keyword)

        # If the keyword is found on the page, store the page number
        if keyword_instances:
            keyword_pages.append(page_number + 1)  # Page numbers start from 1

            # Display the context of the keyword on the page
            for inst in keyword_instances:
                context = page.get_text("text", clip=inst).strip()
                print(f"Page {page_number + 1}, Context: {context}")

    # Close the PDF document
    pdf_document.close()

    # Display the pages where the keyword is found
    print(f"\nKeyword '{keyword}' found on pages: {', '.join(map(str, keyword_pages))}")

# Example usage
pdf_path = "/content/HPP_2021-2022_BurnSurgeAnnex_SCHCC_FINAL.pdf"
search_keyword_in_pdf(pdf_path, "coordination")

```
