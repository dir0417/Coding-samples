```python
import fitz  # PyMuPDF
```

```python

def search_and_extract_description(pdf_path, keyword):
    doc = fitz.open(pdf_path)

    descriptions = []

    for page_number in range(doc.page_count):
        page = doc[page_number]
        text = page.get_text()

        if keyword.lower() in text.lower():
            # Extract a portion of text around the keyword
            start_index = max(text.lower().find(keyword.lower()) - 50, 0)
            end_index = min(text.lower().find(keyword.lower()) + 50, len(text))
            description = text[start_index:end_index]

            descriptions.append({
                'Page': page_number + 1,
                'Description': description.strip()
            })

    doc.close()
    return descriptions

# Example usage:
pdf_path = '/content/HPP_2021-2022_BurnSurgeAnnex_SCHCC_FINAL.pdf'
keyword_to_search = 'coordination'

results = search_and_extract_description(pdf_path, keyword_to_search)

if results:
    print(f"Found {len(results)} occurrences of the keyword '{keyword_to_search}' in the PDF:")
    for result in results:
        print(f"Page {result['Page']}:\n{result['Description']}\n")
else:
    print(f"No occurrences of the keyword '{keyword_to_search}' found in the PDF.")
```
