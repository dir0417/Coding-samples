```python
import PyPDF2
```

```python
def search_and_extract(pdf_file_path, keyword, output_file_path):
    # Open the PDF file
    with open(pdf_file_path, 'rb') as pdf_file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(pdf_file)

        # Initialize an empty string to store the extracted text
        extracted_text = ''

        # Iterate through all pages in the PDF
        for page_number in range(len(pdf_reader.pages)):
            # Extract text from the current page
            page = pdf_reader.pages(page_number)
            page_text = page.extractText()

            # Check if the keyword is present in the page text
            if keyword.lower() in page_text.lower():
                # If found, append the context to the extracted text
                extracted_text += f"Page {page_number + 1}:\n"
                extracted_text += page_text + '\n\n'

        # Save the extracted text to an output file
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(extracted_text)

# Example usage
pdf_file_path = '/content/HPP_2021-2022_BurnSurgeAnnex_SCHCC_FINAL.pdf'
keyword = 'coordination'
output_file_path = '/content/output.txt'

search_and_extract(pdf_file_path, keyword, output_file_path)
```
