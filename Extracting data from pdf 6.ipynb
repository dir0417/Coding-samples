```python
import fitz  # PyMuPDF
from PIL import Image
import pytesseract
from PyPDF2 import PdfReader
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
import pandas as pd
```

```python
# Function to extract text from image using pytesseract
def extract_text_from_image(image_path):
    img = Image.open(image_path)
    text = pytesseract.image_to_string(img)
    return text
```

```python
# Function to extract images from PDF and save as Excel
def extract_images_and_save_to_excel(pdf_path, excel_path):
    pdf = PdfReader(open(pdf_path, "rb"))
    workbook = Workbook()
    sheet = workbook.active

    for page_num in range(pdf.getNumPages()):
        page = pdf.getPage(page_num)
        images = page.get_images(full=True)

        for img_index, img_info in enumerate(images):
            xref = img_info[0]
            base_image = pdf.extract_xobject(xref)
            image_data = base_image["image"]

            image = Image.frombytes(
                image_data["bits"],
                [image_data["width"], image_data["height"]],
                image_data["data"]
            )

            text = extract_text_from_image(image)
            sheet.cell(row=img_index + 1, column=1, value=text)

    workbook.save(excel_path)


# Example usage
if __name__ == "__main__":
    pdf_file = "/content/CDPHE OEPR Supply Chain Integrity Assessment Summary Report - FINAL.pdf"  # Replace with the path to your PDF file
    excel_file = "/content/output.xlsx"      # Replace with the desired Excel output file name

    extract_images_and_save_to_excel(pdf_file, excel_file)
```
